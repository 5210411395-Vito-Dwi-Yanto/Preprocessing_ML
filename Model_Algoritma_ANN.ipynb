# **Menskalakan dan melatih test split**
Sekarang siap untuk melatih model dan memprediksi solusi yang diperlukan

```python
# Features
X = df.drop('price',axis=1)

# Label
y = df['price']

# Split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

# **Normalisasi / penskalaan data**
menskalakan data fitur. Untuk mencegah kebocoran data dari set pengujian, kami hanya menyesuaikan scaler kami dengan set pelatihan.

```python
scaler = MinMaxScaler()      # melakukan penskalaan

# fit and transfrom
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# semuanya telah diskalakan antara 1 dan 0
print('Max: ',X_train.max())
print('Min: ', X_train.min())
```

# **Membuat model**
memperkirakan jumlah neuron (unit) dari fitur kami.

```python
model = Sequential()

# input layer
model.add(Dense(19,activation='relu'))

# hidden layers
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

# output layer
model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x=X_train,y=y_train.values,
          validation_data=(X_test,y_test.values),
          batch_size=128,epochs=400)
```

# **Kerugian pelatihan per zaman**
Plot ini menunjukkan kehilangan pelatihan per zaman.

```python
losses = pd.DataFrame(model.history.history)

plt.figure(figsize=(15,5))
sns.lineplot(data=losses,lw=3)
plt.xlabel('Epochs')
plt.ylabel('')
plt.title('Training Loss per Epoch')
sns.despine()
```

# **Memprediksi data baru**
Pada bagian ini kami memberikan model set pengujian untuk mendapatkan daftar prediksi.

```python
# predictions on the test set
predictions = model.predict(X_test)

print('MAE: ',mean_absolute_error(y_test,predictions))
print('MSE: ',mean_squared_error(y_test,predictions))
print('RMSE: ',np.sqrt(mean_squared_error(y_test,predictions)))
print('Variance Regression Score: ',explained_variance_score(y_test,predictions))

print('\n\nDescriptive Statistics:\n',df['price'].describe())
```

# **Prediksi model vs kecocokan sempurna**
dapat membandingkan prediksi model dengan kecocokan sempurna untuk melihat seberapa akurat model tersebut.

```python
f, axes = plt.subplots(1, 2,figsize=(15,5))

# Our model predictions
plt.scatter(y_test,predictions)

# Perfect predictions
plt.plot(y_test,y_test,'r')

errors = y_test.values.reshape(6484, 1) - predictions
sns.distplot(errors, ax=axes[0])

sns.despine(left=True, bottom=True)
axes[0].set(xlabel='Error', ylabel='', title='Error Histogram')
axes[1].set(xlabel='Test True Y', ylabel='Model Predictions', title='Model Predictions vs Perfect Fit')
```

# **Memprediksi rumah baru**
akan menggunakan model untuk memprediksi harga rumah baru.
